```python
def ext_eps(txt):
    check=False
    data=''
    for i in txt:
        if i==')':
            check=False
            break
        if check==True:
            data+=i
        if i=='(':
            check=True
    return data
```

```python
df['Episodes']=df['Title'].apply(ext_eps)
```

```python
df['Episodes']=df['Episodes'].str.replace('eps','')
```

```python
df['Episodes']=df['Episodes'].astype(int)
```

```python
df['Title'][0]
```

```python
#Make a new column for timespan
def time_span(txt):
    check=False
    data=''
    for i in range(len(txt)):
        if txt[i]==")":
            for j in range(i+1,i+20):
                data+=txt[j]
            return data
```

```python
df['Time_span']=df['Title'].apply(time_span)
```
