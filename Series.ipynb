**Panda Series**

A series is a one dimentional labled array capable of holding any datatype. The axis lables are collectively called the index

**Creative Series**

There are multiple ways to create pandas Series:

```python
lables=['a','b','c']
my_list=[10,20,30]
arr=np.array([10,20,30])
d={1:10,2:20,3:30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list,index=lables)
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
