```python
import numpy as np
import pandas as pd
```

**Find missing data**

```python
data={
    'A':[1,2,np.nan,4,5],
    'B':[np.nan,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5]
}
df=pd.DataFrame(data)
df
```

```python
# isna()will check whether the value is null or not
df.isna()
```

```python
df.isna().sum()
```

```python
#  any() will check whether the whole column has any null value
df.isna().any()
```

**Remove Missing Values**

```python
df.dropna()
```
