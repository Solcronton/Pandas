```python
import numpy as np
import pandas as pd
```

**Find missing data**

```python
data={
    'A':[1,2,np.nan,4,5],
    'B':[1,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5]
}
df=pd.DataFrame(data)
df
```

```python
# isna()will check whether the value is null or not
df.isna()
```

```python
df.isna().sum()
```

```python
#  any() will check whether the whole column has any null value
df.isna().any()
```

**Remove Missing Values**

```python
#It will remove the whole row of having null valur  
df.dropna()
```

```python
df
```

```python
#thresh=n :- here it will select the rows of having n no of non null values 
df.dropna(thresh=2)
```

**Filling the non null values**

```python
df
```

```python
#HEre by using fillna(x) the null values will be replaced with x
#But all this is temperary if you want to change in the main means we want to use implace=True
df.fillna(0)
```

```python
#To give different values to null values in diffent columns use this one 
values={'A':0,'B':100,'C':300,'D':500}
df.fillna(value=values)
```

```python
#To fill the null value with its mean value
df.fillna(df.mean())
```
