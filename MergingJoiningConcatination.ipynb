**Merge**

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

#DataFrame 2: Salary information

salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],'Salary': [60000, 80000, 65000, 70000, 90000],'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
#Here we merge 2 data sets by using a common column 
#inner means it will take the common only
pd.merge(employees,salaries,on='employee_id',how='inner')
```

```python
#outter means it will take the whole employee _id
pd.merge(employees,salaries,on='employee_id',how='outer')
```

```python
#if you use how='left' means it will merge with the bases on 'on' left side value of on
pd.merge(employees,salaries,on='employee_id',how='left')
```

```python
#if you use how='right' means it will merge with the bases on 'on' right side value of on
pd.merge(employees,salaries,on='employee_id',how='right')
```

**COncatination of 2 DataFrames**

```python
df1= pd.DataFrame({
    'A': ['Α0', 'A1', 'A2'], 
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2'] 
})
df2= pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5'] 
})
```

```python
df1
```

**Joning 2 Data Frames**

```python
df1=pd.DataFrame({
    'name':['Alice','Bob','Charlie']
},index=[1,2,3])
df2=pd.DataFrame({
    'Score':[85,95,70]
},index=[2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='outer')
```
